## Dataset can contain missing data, numerical string value, various cues. If we can clean them, we can make easy our analysis.

### #Find Unique Values IN 'Category' Column

```python
df["Category"].unique()
```

##### '1.9' is wrong value. So I have to see in which rows '1.9' corresponds to.

```python
df[df["Category"]=='1.9']
```

##### We can make NaN category value. At that time we can get a clean df. I used shift() method that is found Pandas library.

```python
df.loc[10472] = df.loc[10472].shift()
df['App'].loc[10472] = df['Category'].loc[10472]

df['Category'].loc[10472] = np.nan
df.loc[10472]
```

### #Find Unique Values IN 'Rating' Column

```python
df["Rating"].unique()
```

```python
# Change Datatype Object into float
df["Rating"] = df["Rating"].astype("float")
df["Rating"].dtype
```

### #Find Unique Values IN 'Reviews' Column

```python
df["Reviews"].unique()
```

```python
df["Reviews"] = df["Reviews"].str.replace('3.0M',"3000000")
```

```python
#Change Datatype Object into Float
df["Reviews"] = df["Reviews"].astype("float")
df["Reviews"].dtype
```

### #Find Unique Values IN 'Size' Column

```python
df["Size"].unique()
```

##### Data type of 'Size' Column is object. I had to convert the column because it contains the application sizes. Firstly I changed 'Varies with device' value with Nan. After, I dropped 'M' and 'k'. I changed from '1000+' to 1000. Finally, I converted float value.

```python
df["Size"] = df["Size"].replace('Varies with device',np.NaN)
df["Size"] = df["Size"].str.replace("M","")
df["Size"] = df["Size"].str.replace("k","")
df["Size"] = df["Size"].str.replace("+","")
```

```python
# Change Datatype Object into Float
df["Size"] = df["Size"].astype("float")
df["Size"].dtype
```

### #Find Unique Values IN 'Installs' Column

```python
df["Installs"].unique()
```

##### Data type of 'Installs' Column is object. I'm gonna make similar processes, which I made the in 'Size' Column.

```python
df["Installs"] = df["Installs"].str.replace("+","")
df["Installs"] = df["Installs"].str.replace(",","")
```

```python
# Change Datatype Object into Float
df["Installs"] = df["Installs"].astype("float")
df["Installs"].dtype
```

### #Find Unique Values IN 'Price'Column

```python
df["Price"].unique()
```

##### Data type of 'Price' Column is object. I'm gonna make similar processes, which I made the in 'Installs' Column.

```python
df["Price"] = df["Price"].str.replace("$","")
```

```python
df["Price"] = df["Price"].astype("float")
df["Price"].dtype
```

### #Find Unique Values IN 'Last Updated' Column

```python
df["Last Updated"].unique()
```

##### Data type of "Last Uptated" Column is object. I convert from object to datetime type.

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])
df["Last Updated"]
```

# EXPLORATORY DATA ANALYSIS

### #Word Cloud

```python
plt.figure(figsize=(12,7))
wordcloud = WordCloud(background_color="white", width=512,height=384).generate(" ".join(df))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

### #Find Correlation Map

```python
plt.figure(figsize=(12,7))
sns.heatmap(data=df.corr(),annot=True)
plt.show()
```

### Distribution Of Rating 

```python
plt.figure(figsize=(15,7))
ax = sns.countplot(data=df,x=df["Rating"])
plt.title(" Distribution Of Rating ", weight = "bold",fontsize = 18,pad=10)
plt.xticks(weight="bold",rotation=90)
plt.yticks(weight="bold")
plt.xlabel("Rating",weight="bold",labelpad = 12,fontsize = 12)
plt.ylabel("Count",weight="bold",labelpad = 12,fontsize = 12)

for i in ax.containers:
    i.datavalues
    ax.bar_label(i,fontweight="bold")
    
plt.show()
```
