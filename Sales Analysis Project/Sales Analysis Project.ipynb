## ***What was the best month for sales? How much was earned that month?***

### Add month Column

#### To add month column , all we need to do is to use date time methods.

```python
Sales_Data["month_name"] = Sales_Data["Order Date"].dt.month_name()
Sales_Data["month"] = Sales_Data["Order Date"].dt.month
Sales_Data.head(3)
```

#### In This DataFrame , There are Two columns 'Price Each' and 'Quantity Ordered'.  But we do not have Total Sale Value (TSV), We should also add a calculated column Total Sale Value. Let's do it first.

```python
Sales_Data["Total Sale Value"] = Sales_Data["Quantity Ordered"]*Sales_Data["Price Each"]
Sales_Data.head(5)
```

### Now we can easily use groupby method to find maximum sales value.

```python
Best_month_Sale = Sales_Data.groupby(["month","month_name"])["Total Sale Value"].sum()
Best_month_Sale = Best_month_Sale.reset_index()
Best_month_Sale
```

```python
## Best month Sale is-
Best_month_Sale.loc[Best_month_Sale["Total Sale Value"] == Best_month_Sale["Total Sale Value"].max()]
```

```python
z =Best_month_Sale["Total Sale Value"].max()
print(f"Total Sale Value In December: {z}")
```

##### December is a Festival month , the biggest thing is December is Christmas, So people spend money and manby that is the reason for december is best month for sales.

### Plot The Data

```python
plt.figure(figsize=(12,6))
sns.barplot(data=Best_month_Sale,x="month_name",y="Total Sale Value")
plt.xticks(weight="bold")
plt.yticks(weight="bold")
plt.xlabel("month_name",weight = "bold",labelpad=10,fontsize=12)
plt.ylabel("Total Sale Value",weight = "bold",labelpad=10,fontsize=12)
plt.show()
```

##  ***What products are most often sold together?***

#### This is going to be a tricky business. Before diving into the problem, let us understand the problem.
#### What Products are most often sold together gives a very important insight that helps company lure customers into spending more and generate more profit.
#### You must have seen Amazon's Customer Who Bought this item also bought this Item. We are going to do something like this.
#### Let's try to solve this problem Step by step : 

### Duplicate Order ID : 

#### Same Order ID for different products suggests customer bought different products Together in Our Data. Let us look for Duplicate OrderId Products: 

```python
# All Duplicate Order ID 
Duplicate_Order = Sales_Data.loc[Sales_Data["Order ID"].duplicated(keep=False)]
Duplicate_Order
```

### Group Different Products With Same Order ID

#### Let's create a new column that contains different Products with the same Order ID. We can do this with the help of transform method.

```python
Duplicate_Order["Grouped"] = Duplicate_Order.groupby("Order ID")["Product"].transform(lambda x:",".join(x))
Duplicate_Order
```

#### counting-unique-pairs-of-numbers-

```python
Duplicate_Order.drop_duplicates(subset="Order ID",inplace=True)
Duplicate_Order
```

```python
Duplicate_Order = Duplicate_Order["Grouped"].value_counts().reset_index()
Duplicate_Order.rename(columns={"index":"Products","Grouped":"Products Count"},inplace=True)
Duplicate_Order
```

```python
Duplicate_Order = Duplicate_Order.head(10)
Duplicate_Order
```

```python
m = Duplicate_Order["Products"]
n = Duplicate_Order["Products Count"]
```

### Plot The Data

```python
plt.figure(figsize=(12,6))
plt.bar(m,n)
plt.xticks(rotation=90 ,weight="bold")
plt.yticks(weight="bold")
plt.xlabel("Products",weight = "bold",fontsize=15)
plt.ylabel("Products Count",weight = "bold",labelpad=10,fontsize=12)

for i in range(len(m)):
    plt.annotate(f"{n[i]}",(m[i],n[i]),weight="bold",ha="center",va="bottom",color="red")
plt.show()
```
