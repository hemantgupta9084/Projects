# DATA PREPROCESSING 

#### #Drop Unnecessary Columns 

```python
empty_list = []
for i in df.columns:
    if i.startswith("FLAG"):
        empty_list.append(i)
empty_list
```

```python
df.drop(columns=empty_list,inplace=True)
df
```

#### #Dealing With Missing Values (Columns 10 %)

```python
empty_col = df.isna().sum()
empty_col = empty_col.loc[empty_col.values>df.shape[0]*0.10]
df = df.drop(columns=empty_col.index)
df
```

#### #Dealing With Missing Values (Rows 10 %)

```python
df = df.dropna(thresh=df.shape[1]-df.shape[1]*0.10)
df
```

### #Dealing With XNA Gender

```python
df["CODE_GENDER"].value_counts()
```

```python
df.loc[df["CODE_GENDER"] == "XNA","CODE_GENDER"] = "F"
df["CODE_GENDER"].value_counts()
```

### #Dealing With Organization Type

```python
df["ORGANIZATION_TYPE"].value_counts().head()
```

```python
df = df.loc[df["ORGANIZATION_TYPE"]!="XNA"]
df
```

### #Finding Outliers And Deleting Them In Income Column

```python
q1,q2,q3 = np.percentile(df["AMT_INCOME_TOTAL"],[25,50,75])
iqr = q3-q1
Lower_extreme = q1-1.5*iqr
Upper_extreme = q3+1.5*iqr
df = df.loc[(df["AMT_INCOME_TOTAL"]<=Upper_extreme) & (df["AMT_INCOME_TOTAL"]>=Lower_extreme)]
df
```

#### #Creating Income Range

```python
bins = np.arange(0,df["AMT_INCOME_TOTAL"].max()+25000,25000)

slots = ['0-25000', '25000-50000','50000-75000','75000-100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
       '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-Above']

df["AMT_INCOME_RANGE"]= pd.cut(df["AMT_INCOME_TOTAL"],bins=bins,labels=slots)
df
```
